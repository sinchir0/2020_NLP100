# 16. ファイルをN分割する
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．

```python
!split -l 695 popular-names.txt popular-names_split
```

```python
# ans16.pyの中身
import sys
import pandas as pd

if __name__ == '__main__':
    
    args = sys.argv

    n = int(args[1])
    
    df = pd.read_csv('popular-names.txt', sep='\t', header=None)
    split_point = int(df.shape[0] / n)
    
    for i in range(n):
        split_num = i + 1
        if (split_num) == 1:
            df[:split_point].to_csv(f'ans16_{split_num}_per_{n}.txt', sep='\t', index=False, header=None)
        elif (split_num != 1 & split_num != n):
            df[split_point*(split_num - 1) : split_point*(split_num)].to_csv(f'ans16_{split_num}_per_{n}.txt', sep='\t', index=False, header=None)
        elif split_num == n:
            df[split_point*(split_num -1):].to_csv(f'ans16_{split_num}_per_{n}.txt', sep='\t', index=False, header=None)
```
